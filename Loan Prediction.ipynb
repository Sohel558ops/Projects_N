### OUTLIERS ANALYSIS

```python
Q1= np.percentile(loan_df['ApplicantIncome'],25)
Q2= np.percentile(loan_df['ApplicantIncome'],50)
Q3=np.percentile(loan_df['ApplicantIncome'],75)

IQR= Q3-Q1
lb= Q1-1.5*IQR
ub= Q3+1.5*IQR
L1= []
median = loan_df['ApplicantIncome'].median()
for i in loan_df['ApplicantIncome']:
    if i<lb or i>ub:   #condition for outliers
        L1.append(median)  #handling outliers by replacing with median values
    else:
        L1.append(i)

loan_df['ApplicantIncome_new']=L1
plt.boxplot(loan_df['ApplicantIncome_new'])
plt.show()        
```

```python
Q1= np.percentile(loan_df['CoapplicantIncome'],25)
Q2= np.percentile(loan_df['CoapplicantIncome'],50)
Q3=np.percentile(loan_df['CoapplicantIncome'],75)

IQR= Q3-Q1
lb= Q1-1.5*IQR
ub= Q3+1.5*IQR
L1= []
median = loan_df['CoapplicantIncome'].median()
for i in loan_df['CoapplicantIncome']:
    if i<lb or i>ub:   #condition for outliers
        L1.append(median)  #handling outliers by replacing with median values
    else:
        L1.append(i)

loan_df['CoapplicantIncome_new']=L1
plt.boxplot(loan_df['CoapplicantIncome_new'])
plt.show()        
```

### MULTI-VARIATE ANALYSIS

```python
path=r'C:\Users\ASUS\Naresh_IT\DATA SCIENCE\Data files\train_ctrUa4K.csv'
loan_df=pd.read_csv(path)

num_cols = loan_df.select_dtypes(exclude='object').columns
for i in num_cols:
    loan_df[i].fillna(loan_df[i].mean(), inplace=True)

plt.figure(figsize=(10,8))
corr_loan = loan_df.corr(numeric_only= True)
sns.heatmap(corr_loan,annot=True)
plt.show()
```

#### INSIGHTS FROM HEAT-MAP:

##### From the Heat map, we can infer the following:
    
    
    #####-The 'LoanAmount' has a strong positive correlation to 'ApplicantIncome' 0.57

    #####-'ApplicanIncome' has correlated to 'Coapplicant Income','Loan Amount Term', and 'Cedit History' negatively

    #####-Higher the income of the applicant, higher will be the amount of loan approved ,hence both are positively related.
    ##### As the applican is highly capable of repaying the loan, the coapplicant need not be a high earner and 
    ##### the loan term for paying the loan back is lower, hence these have negative correlation with the Applicant Income.
